## Block 2: Calling an LLM from Python

### Ollama Quick Reference
- **What is Ollama?** A server running AI models locally on your laptop.
- **Check it's installed:** `ollama --version`
- **Check models:** `ollama list`
- **Start server (if not running):** `ollama serve` (leave terminal open)
- **Address:** `localhost:11434`
  - `localhost` = your own computer (not the internet)
  - `11434` = port number (like a door Ollama listens behind)

```python
import requests
import json
```

```python
def ask_llm(prompt):
    response = requests.post(
        "http://localhost:11434/api/chat",
        json={
            "model": "qwen3:1.7b",
            "messages": [
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            "stream": False
        }
    )
    result = response.json()["message"]["content"]
    if "<think>" in result:
        result = result.split("</think>")[-1].strip()
    return result
```

```python
answer = ask_llm("What is Python in one sentence?")
print(answer)
```

```python
answer = ask_llm("Explain gravity to a 5-year-old")
print(answer)
```

## Block 3: Building a Chatbot with Memory

```python
# Without memory: the model doesn't know what "there" means
print(ask_llm("What is the capital of France?"))
print(ask_llm("What language do they speak there?"))
```

> **Note:** If running the chatbot crashes your notebook (kernel dies), it may be a memory issue with Ollama.
> **Fix:** Restart Ollama ( in a fresh terminal), then restart your notebook kernel and try again.
> The Qwen 1.7B model needs ~2GB RAM free while running.

```python
def chat():
    messages = [
        {"role": "system", "content": "You are a friendly AI assistant. Keep responses to 2-3 sentences."}
    ]
    print("Chatbot ready! Type 'quit' to exit.\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() == "quit":
            print("Goodbye!")
            break

        messages.append({"role": "user", "content": user_input})

        response = requests.post(
            "http://localhost:11434/api/chat",
            json={
                "model": "qwen3:1.7b",
                "messages": messages,
                "stream": False
            }
        )

        assistant_message = response.json()["message"]["content"]
        if "<think>" in assistant_message:
            assistant_message = assistant_message.split("</think>")[-1].strip()

        messages.append({"role": "assistant", "content": assistant_message})
        print(f"AI: {assistant_message}\n")
```

```python
# Run the chatbot! Try:
# 1. "What is the capital of France?"
# 2. "What language do they speak there?"
# 3. "Give me a fun fact about that country"
# Type 'quit' to exit.
chat()
```

## Experimentation

Try changing the system prompt and temperature:

```python
# Pirate mode
def ask_llm_custom(prompt, system_prompt, temperature=0.7):
    response = requests.post(
        "http://localhost:11434/api/chat",
        json={
            "model": "qwen3:1.7b",
            "messages": [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": prompt}
            ],
            "stream": False,
            "options": {"temperature": temperature}
        }
    )
    result = response.json()["message"]["content"]
    if "<think>" in result:
        result = result.split("</think>")[-1].strip()
    return result

print(ask_llm_custom("What is Python?", "You are a pirate captain. Respond in pirate speak."))
```

```python
# Temperature comparison
print("Temperature 0 (predictable):")
print(ask_llm_custom("Give me a startup idea", "Be creative.", temperature=0.0))
print()
print("Temperature 1 (creative):")
print(ask_llm_custom("Give me a startup idea", "Be creative.", temperature=1.0))
```
